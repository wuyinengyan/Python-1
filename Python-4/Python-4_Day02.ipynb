# 4.7 随机漫步

```python
# Python实现
import random

position = 0
walk = [position]
steps = 1000

for i in range(steps):
    step = 1 if random.randint(0, 1) > 0 else -1
    position += step
    walk.append(position)
        
import matplotlib.pyplot as plt

plt.plot(walk[:100])
```

```python
import numpy as np

nsteps = 1000
draws = np.random.randint(0, 2, size=nsteps)  # [low, high)，和Python的不同
steps = np.where(draws > 0, 1, -1)
walk = steps.cumsum()

import matplotlib.pyplot as plt

plt.plot(walk[:100])
```

```python
walk[:100].min(), walk[:100].max()
```

```python
arr_abs = np.abs(walk[:100]) >= 10  # 是否朝某个方向累计走了10距离
arr_abs, arr_abs.argmax()  # 第一次朝某个方向累计走了10的距离
```
